```python
import pandas as pd
import numpy as np
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
path_to_data = "/home/yhbedoya/Datathon/reviews_master_parquet/"
categories = ["All Electronics"]
```

```python
df_list = []
for category in categories:
    catDf = pd.read_parquet(f'{path_to_data}main_cat={category}/', engine="pyarrow")
    catDf['main_cat'] = category
    df_list.append(catDf)

df = pd.concat(df_list)
df
```

```python
df["brand"].value_counts()
```

```python
df_panasonic = df[df["brand"] == "Samsung"]
```

```python
column_counts = df_panasonic["title"].value_counts()
unique_values = column_counts.index
counts = column_counts.values
```

```python
BestProduct = unique_values[0]
df_prod= df_panasonic[df_panasonic["title"] == BestProduct]
df_prod
```

```python
df_prod["vote"].value_counts()
```

```python
df_best = df_prod[df_prod["vote"] > 500]
df_best
```

```python
len(df_best["reviewText"].iloc[0])
```

```python
reviews = "\n".join(df_best["reviewText"].to_list())
len(reviews)
```

```python
review = reviews
```

```python
from transformers import BartTokenizer, BartForConditionalGeneration

# Load the pre-trained BART model and tokenizer
model_name = "facebook/bart-large-cnn"
tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)
```

```python
# Tokenize the review and perform the summarization
inputs = tokenizer.encode("summarize: " + review, return_tensors="pt", max_length=1024, truncation=True)
summary_ids = model.generate(inputs, max_length=150, min_length=50, length_penalty=2.0, num_beams=4, early_stopping=True)

# Decode and print the summary
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print("Summary:", summary)
```
